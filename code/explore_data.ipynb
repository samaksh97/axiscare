```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv("COVID-19_Reported_Patient_Impact_and_Hospital_Capacity_by_State_Timeseries.csv")
```

```python
raw_data["date"] = pd.to_datetime(raw_data["date"])
raw_data.dtypes
```

```python
raw_data = raw_data.sort_values(by=["state","date"]).reset_index(drop=True)
pd.options.display.max_columns=None
raw_data["critical_staffing_shortage_today_yes_perc"] = raw_data["critical_staffing_shortage_today_yes"]/(raw_data["critical_staffing_shortage_today_yes"]	+raw_data["critical_staffing_shortage_today_no"]	)

raw_data.head(3)
```

```python
state = "CA"
state_data = raw_data.loc[raw_data["state"]==state]
#state_data = raw_data
state_data["critical_staffing_shortage_today_yes_perc"] = state_data["critical_staffing_shortage_today_yes"]/(state_data["critical_staffing_shortage_today_yes"]	+state_data["critical_staffing_shortage_today_no"]	)
state_data["inpatient_beds_perc"] = state_data["inpatient_beds"]/state_data["inpatient_beds_coverage"]
state_data["inpatient_beds_used_covid_perc"] = state_data["inpatient_beds_used_covid"]/state_data["inpatient_beds_used_covid_coverage"]
```

```python
import plotly.express as px
fig = px.line(state_data, x="date", y="inpatient_beds_perc", title='Inpatient beds of California',markers=True)
fig.show()
```

```python
cases_data = pd.read_csv("D:/Hacklytics/us_cases_state_clean.csv").drop("Unnamed: 0",axis=1)
cases_data["date"] = pd.to_datetime(cases_data["date"],infer_datetime_format=True)
cases_data = cases_data.sort_values(by=["usa_state_code","date"]).reset_index(drop=True)
cases_data["daily_deaths"] = cases_data.groupby("usa_state_code")["deaths"].diff(1)
cases_data["daily_cases"] = cases_data.groupby("usa_state_code")["cases"].diff(1)
```

```python
state = "CA"
state_data = raw_data.loc[raw_data["state"]==state]
#state_data = raw_data
state_data["critical_staffing_shortage_today_yes_perc"] = state_data["critical_staffing_shortage_today_yes"]/(state_data["critical_staffing_shortage_today_yes"]	+state_data["critical_staffing_shortage_today_no"]	)
state_data["inpatient_beds_perc"] = state_data["inpatient_beds"]/state_data["inpatient_beds_coverage"]
state_data["inpatient_beds_used_covid_perc"] = state_data["inpatient_beds_used_covid"]/state_data["inpatient_beds_used_covid_coverage"]
state_data["week"] = state_data.date.dt.week
state_data["year"] = state_data.date.dt.year
def pad(x):
    if x<10:
        return "0"+str(x)
    else:
        return str(x)
state_data["year_week"] = state_data["year"].astype(str)+"_"+state_data["week"].apply(lambda x: pad(x))

state_data = state_data.groupby("year_week",as_index=False)["critical_staffing_shortage_today_yes_perc"].mean().rename(columns={"critical_staffing_shortage_today_yes_perc":"avg_critical_staffing_shortage_today_yes_perc"})

#cases_data = pd.read_csv("D:/Hacklytics/us_cases_state_clean.csv").drop("Unnamed: 0",axis=1)
#cases_data["date"] = pd.to_datetime(cases_data["date"],infer_datetime_format=True)
#cases_ca =cases_data.loc[cases_data["usa_state_code"]==state]
cases_ca = cases_data.loc[cases_data["usa_state_code"]==state]
cases_ca["week"] = cases_ca.date.dt.week
cases_ca["year"] = cases_ca.date.dt.year
cases_ca["year_week"] = cases_ca["year"].astype(str)+"_"+cases_ca["week"].apply(lambda x: pad(x)) 
cases_ca = cases_ca.groupby("year_week",as_index=False)["daily_deaths"].sum().rename(columns={"daily_deaths":"weekly_deaths"})
cases_ca.head(2)
whole_data = state_data.merge(cases_ca,on="year_week",how="left")
#cases_ca["daily_deaths"] = cases_ca["deaths"] - cases_ca["deaths"].shift(1).fillna(0)
whole_data.head(5)
```

```python
fig = px.line(cases_ca, x="year_week", y="weekly_deaths", title="COVID-19 Deaths per week in "+state,markers=True)
#fig.add_scatter(x=state_data['date'], y=state_data['inpatient_beds'])
fig.show()
fig.write_html("covid_19_weekly_deaths_california.html")
```
